```python
# combine df2 and df3 by concatenation to avoid NaNs and keep feature names unchanged
df23 = pd.concat([df2, df3], ignore_index=True).drop_duplicates()

# save merged dataframe
output_dir = Path().resolve().parent / "data/processed"
os.makedirs(output_dir, exist_ok=True)
df23.to_csv(output_dir / "households_coaching_visits_merged.csv", index=False)
```

```python
# Create a clear summary table with named columns and rows
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Total households per region, district, cluster (from df1)
households_summary = df1.groupby(['region_name', 'district_name', 'cluster_name'], as_index=False)['household_id'].nunique()
households_summary = households_summary.rename(columns={'household_id': 'total_households'})

# 2. Number of households attended each Training Type (from df4)
training_attendance = df4.groupby(['region_name', 'district_name', 'cluster_name', 'training_type'], as_index=False)['household_id'].nunique()
training_attendance = training_attendance.pivot_table(index=['region_name', 'district_name', 'cluster_name'], columns='training_type', values='household_id', fill_value=0)
training_attendance = training_attendance.reset_index()

# 3. Proportion of households attended each training type
for col in training_attendance.columns[3:]:
    training_attendance[col + '_prop'] = training_attendance[col] / households_summary['total_households']

# 4. Proportion of households visited once, twice, thrice (from df23)
visit_counts = df23.groupby(['region_name', 'district_name', 'cluster_name', 'household_id']).size().reset_index(name='visit_count')
visit_counts['visit_count'] = visit_counts['visit_count'].astype(int)
for n in [1,2,3]:
    visit_counts[f'visited_{n}'] = (visit_counts['visit_count'] == n).astype(int)
visit_summary = visit_counts.groupby(['region_name', 'district_name', 'cluster_name'])[[f'visited_{n}' for n in [1,2,3]]].sum().reset_index()
for n in [1,2,3]:
    visit_summary[f'visited_{n}_prop'] = visit_summary[f'visited_{n}'] / households_summary['total_households']

# Merge all summaries into a single table
summary = households_summary.merge(training_attendance, on=['region_name', 'district_name', 'cluster_name'], how='left')
summary = summary.merge(visit_summary, on=['region_name', 'district_name', 'cluster_name'], how='left')

# Reorder columns for clarity
cols = ['region_name', 'district_name', 'cluster_name', 'total_households']
if len(training_attendance.columns) > 3:
    cols += [col for col in training_attendance.columns if col not in cols and not col.endswith('_prop')]
    cols += [col for col in training_attendance.columns if col.endswith('_prop')]
cols += [f'visited_{n}' for n in [1,2,3]]
cols += [f'visited_{n}_prop' for n in [1,2,3]]
summary = summary[cols]

# Output as dataframe
summary.fillna(0, inplace=True)
display(summary)

# Save as PNG
os.makedirs(Path().resolve().parent / 'visuals', exist_ok=True)
plt.figure(figsize=(min(24, 4+2*len(summary.columns)), min(8, 0.5*len(summary))))
sns.heatmap(summary.set_index(['region_name', 'district_name', 'cluster_name']).astype(float), annot=True, fmt='.2f', cmap='Blues')
plt.title('Summary Table Metrics')
plt.tight_layout()
plt.savefig(Path().resolve().parent / 'visuals' / 'summary_table.png')
plt.close()
```

### Key Challenges & their Solutions


1. **Inconsistent Household Identifiers:** Different datasets used different column names for household IDs (e.g., `household_id` vs `hhid`).
   - *Solution:* Inspected and standardized the column names, dropping duplicates and unnecessary columns.


2. **Combining Visit Datasets Without NaNs:** Merging with outer joins introduced many NaNs due to non-overlapping columns.
   - *Solution:* Used `pd.concat` to stack the visit datasets vertically, preserving all records and columns, then dropped duplicates.


3. **Aligning Groupby and Pivot Results:** Ensuring that all summary tables (counts, proportions, visits) had matching keys and clear column names for merging and final output.
   - *Solution:* Used `reset_index()` and explicit column ordering to guarantee merge compatibility and clarity in the summary table.


4. **Disaggregating by Multiple Levels:** Required grouping by region, district, and cluster for all metrics.
   - *Solution:* Used pandas groupby and pivot_table to aggregate and reshape data as needed.


5. **Calculating Proportions with Varying Denominators:** Needed to ensure proportions were calculated using the correct total households per group.
   - *Solution:* Merged summary tables carefully and used the total from the household list as the denominator.


6. **Exporting Data as PNG:** Dataframes can be large and hard to visualize as images.
   - *Solution:* Used seaborn heatmap for a compact, annotated visual summary and saved the output to the visuals folder.

```python
# convert startTime & endTime to readable datetime if they are Excel serial numbers
import pandas as pd
import datetime

def excel_serial_to_datetime(series):
    # Handle NaNs and non-numeric values gracefully
    return pd.to_datetime('1899-12-30') + pd.to_timedelta(pd.to_numeric(series, errors='coerce'), unit='D')

df23['startTime_dt'] = excel_serial_to_datetime(df23['startTime'])
df23['endTime_dt'] = excel_serial_to_datetime(df23['endTime'])

df23[['startTime', 'startTime_dt', 'endTime', 'endTime_dt']].head(10)
```

```python
# Fill nulls in timeTakenToCompleteFormInMinutes with (endTime_dt - startTime_dt) in minutes
mask = df23['timeTakenToCompleteFormInMinutes'].isnull()
df23.loc[mask, 'timeTakenToCompleteFormInMinutes'] = (df23.loc[mask, 'endTime_dt'] - df23.loc[mask, 'startTime_dt']).dt.total_seconds() / 60
df23['timeTakenToCompleteFormInMinutes'].head(10)
```

```python
# drop original startTime and endTime columns

df23 = df23.drop(columns=["startTime", "endTime"])

# drop nulls subseting on household_id
df23 = df23.dropna(subset=['household_id'])
```

- Dropped startTime & endTime features which are now irrelevant.
- Since for some observations there were no household_id observations, i dropped nulls subsetting on household_id so that i work with only known household_ids

```python

# columns for WASH (H:AB), Agriculture (AC:AP), VSLA Participation
wash_cols = df23.columns[7:28]  # H:AB (0-based index)
agri_cols = df23.columns[28:42]  # AC:AP
vsla_cols = [col for col in df23.columns if 'vsla' in col.lower() or 'VSLA' in col]

# Computing mean for each household
df23['wash_score'] = df23[wash_cols].mean(axis=1, skipna=True)
df23['agri_score'] = df23[agri_cols].mean(axis=1, skipna=True)
df23['vsla_score'] = df23[vsla_cols].mean(axis=1, skipna=True) if vsla_cols else np.nan

# Computing overall average adoption score
df23['overall_adoption_score'] = df23[['wash_score', 'agri_score', 'vsla_score']].mean(axis=1, skipna=True)

# Aggregate by household_id
household_scores = df23.groupby('household_id')[['wash_score', 'agri_score', 'vsla_score', 'overall_adoption_score']].mean().reset_index()

household_scores.head()
```

```python
# Using the first visit per household 
first_visits = df23.sort_values(['household_id', 'startTime_dt']).groupby('household_id', as_index=False).first()

# coaching components columns H:AP (0-based: 7:42)
coaching_cols = df23.columns[7:42]
score_cols = ['wash_score', 'agri_score', 'vsla_score', 'overall_adoption_score']

# Aggregate by Region, District, Cluster
agg_cols = list(coaching_cols) + score_cols
adoption_summary = first_visits.groupby(['region_name', 'district_name', 'cluster_name'])[agg_cols].mean().reset_index()

adoption_summary.head()
```

### My Comments on Adoption Scores by Region, District, and Cluster

- **WASH Performance:**
  - Regions, districts, or clusters with higher average WASH scores indicate better adoption of water, sanitation, and hygiene practices. Look for clusters with consistently high or low scores to identify best practices or areas needing intervention.

- **Agriculture Performance:**
  - High agriculture scores reflect strong uptake of agricultural coaching components. Districts or clusters with lower averages may benefit from targeted support or further training.

- **VSLA Participation:**
  - VSLA (Village Savings and Loan Association) participation scores show the extent of financial inclusion and group savings adoption. Regions with low VSLA scores may have barriers to participation or require more sensitization.

- **Overall Adoption:**
  - The overall adoption score provides a holistic view, combining WASH, Agriculture, and VSLA. Regions or clusters with high overall scores are likely more resilient and have embraced the coaching program broadly. Conversely, areas with low overall scores may need a comprehensive review to identify and address adoption barriers.
